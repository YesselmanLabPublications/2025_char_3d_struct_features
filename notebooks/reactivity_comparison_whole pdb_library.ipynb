```python
import pandas as pd
import matplotlib.pyplot as plt
from dms_3d_features.plotting import plot_whole_pdb_reactivity, publication_style_ax, plot_motif_boxplot_stripplot
```

```python
DATA_PATH = "../dms_3d_features/resources/jsons"
```

```python
df = pd.read_json(f'{DATA_PATH}/dms_comparison_with_z_score1.json')
```
